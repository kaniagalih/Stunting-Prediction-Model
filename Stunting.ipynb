```python
#encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in data.columns:
  if data[column].dtype == 'object':
    data[column + '-Encoded'] = le.fit_transform(data[column])

#Delete data kategori
data_encoded = data.drop(columns=data.select_dtypes(include='object').columns)
print(data_encoded)
```

Hasil Encoding

Male -> 1

Female -> 0

No -> 0

Yes -> 1

```python
#dependent & Independent
X = data_encoded.iloc[:, :-1].values
y = data_encoded.iloc[:, -1]
```

```python
X
```

```python
y
```

```python
#Normalize
from sklearn.preprocessing import Normalizer
transform = Normalizer().fit(X)
transform
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
scaler.transform(X)
```

```python
#Optimization using PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(X)
```

```python
#Split 70 30
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# #Undersampling
# from imblearn.under_sampling import RandomUnderSampler
# rus = RandomUnderSampler(random_state=42)
# rus.fit_resample(X_train, y_train)
```

```python
#Oversampling
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
smote.fit_resample(X_train, y_train)
```

#Model

Random Forest

```python
#Model Random Forest
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(max_depth=5, random_state=0)
RF.fit(X_train, y_train)
```

```python
#Predict Random Forest
pred_RF = RF.predict(X_test)
pred_RF
```

```python
#Evaluation Random Forest
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
cf_RF = classification_report(y_test, pred_RF, labels=[0, 1])
acc_RF = accuracy_score(y_test, pred_RF)
print(cf_RF)
print('Accuracy : ', acc_RF)
```

```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cm_RF = confusion_matrix(y_test, pred_RF)
print(cm_RF)
```

Logistic Regression

```python
#Model
from sklearn.linear_model import LogisticRegression
Logreg = LogisticRegression(random_state = 0)
Logreg.fit(X_train, y_train)
```

```python
#Predict Logistic Regression
pred_Logreg = Logreg.predict(X_test)
pred_Logreg
```

```python
#Evaluation
cf_LR = classification_report(y_test, pred_Logreg, labels=[0, 1])
acc_LR = accuracy_score(y_test, pred_Logreg)
print(cf_LR)
print('Accuracy : ', acc_LR)
```

```python
#Confusion matrix
cm_LR = confusion_matrix(y_test, pred_Logreg)
print(cm_LR)
```
